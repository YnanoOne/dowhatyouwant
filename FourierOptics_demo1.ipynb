```python
from FourierOptics.utils import *
```

```python
# all in um
lam = 0.522     
z = 5         
N = 1000
dx = 1e-2      

x, X, Y = make_squareGrid(N, dx)
```

```python
r = 0.5/2  # um
circ = ((X**2 + Y**2) <= r**2).astype(float)

U2 = fresnel_propagate(circ, lam=lam, z=z, dx=dx)
I2 = np.abs(U2)**2
I2 /= I2.max()

plt.figure()
plt.imshow(I2, origin="lower")
plt.xlabel("x (um)")
plt.ylabel("y (um)")
plt.title(f"Fresnel Propagate z= {z}")
plt.colorbar()
plt.show()
```

```python
r = 0.5/2  # um
circ = ((X**2 + Y**2) <= r**2).astype(float)

U2 = helmholtz_propagate(circ, lam=lam, z=z, dx=dx)
I2 = np.abs(U2)**2
I2 /= I2.max()

plt.figure()
plt.imshow(I2, origin="lower")
plt.xlabel("x (um)")
plt.ylabel("y (um)")
plt.title(f"Fresnel Propagate z= {z}")
plt.colorbar()
plt.show()
```
